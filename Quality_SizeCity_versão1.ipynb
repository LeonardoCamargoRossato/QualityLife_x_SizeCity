# Classificando por quartis (os valores des colunas)

```python
def mapeia_quatis(coluna):
    nova_col = []
    for i in coluna:
        if i <= coluna.quantile(0.25):
            nova_col.append(1)
        elif i <= coluna.quantile(0.50) and i > coluna.quantile(0.25):
            nova_col.append(4)
        elif i <= coluna.quantile(0.75) and i > coluna.quantile(0.50):
            nova_col.append(8)
        else:
            nova_col.append(16)

    df_BR_IDHM['Quartis'] = nova_col
```

```python
mapeia_quatis(df_BR_IDHM.Pop_2018)
```

## Plotando gráfico: Fit log_x  --> IDHM_2010 x Pop_2018_City(AP)

```python
def plota(axis_x,axis_y,trendline,tam):
    if trendline == 'log':
        tipo_de_ajuste = dict(log_x=True)
    elif trendline == 'reta':
        tipo_de_ajuste = dict(log_x=False)
    fig = px.scatter(df_BR_IDHM, x=axis_x, y=axis_y, marginal_x="box", marginal_y="box", 
                     trendline="ols", trendline_options=tipo_de_ajuste, trendline_color_override="red",
                     template="simple_white", hover_name='Municipio', size=tam,
                     height=500)
    fig.show()
```

```python
colunas = ['IDHM_2010','IDHM_Renda_2010','IDHM_Longevidade_2010','IDHM_Educação_2010','Pop_2018','Area','pop_por_area','Quartis_Area','Quartis']

interact(plota,axis_x=colunas,axis_y=colunas, trendline=['log','reta'],tam=colunas)
```
